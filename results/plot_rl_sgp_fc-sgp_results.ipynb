```python
import matplotlib.pyplot as plt
import glob
from astropy.io import fits
```

```python
from mpl_toolkits.axes_grid1 import make_axes_locatable

def axes_colorbar(ax):
    divider = make_axes_locatable(ax)
    cax = divider.append_axes('right', size='5%', pad=0.05)
    return cax
```

```python
i = 0
fig, ax = plt.subplots(10, 4, figsize=(15, 30))
for f in zip(sorted(glob.glob('FC_SGP_original_images/*.fits')), sorted(glob.glob('RL_reconstructed_images/*.fits')), sorted(glob.glob('SGP_reconstructed_images/*.fits')), sorted(glob.glob('FC_SGP_reconstructed_images/*.fits'))):
    if f[0].split('/')[1] == 'ccfbvc170119.fits_515.4571_110.0911_SGP_orig.fits' or f[0].split('/')[1] == 'ccfbvc170121.fits_178.0998_506.4224_SGP_orig.fits' or f[0].split('/')[1] == 'ccfbvc170119.fits_226.8246_174.0645_SGP_orig.fits':
        continue

    if i == 10:
        break

    if i == 0:
        ax[i, 0].set_title("Original")
        ax[i, 1].set_title("RL")
        ax[i, 2].set_title("SGP")
        ax[i, 3].set_title("FC-SGP")

    orig = fits.getdata(f[0])
    rl = fits.getdata(f[1])
    sgp = fits.getdata(f[2])
    fcsgp = fits.getdata(f[3])
    
    im0 = ax[i, 0].imshow(orig)
    cax0 = axes_colorbar(ax[i, 0])
    fig.colorbar(im0, cax=cax0, orientation='vertical')

    im1 = ax[i, 1].imshow(rl)
    cax1 = axes_colorbar(ax[i, 1])
    fig.colorbar(im1, cax=cax1, orientation='vertical')

    im2 = ax[i, 2].imshow(sgp)
    cax2 = axes_colorbar(ax[i, 2])
    fig.colorbar(im2, cax=cax2, orientation='vertical')

    im3 = ax[i, 3].imshow(fcsgp)
    cax3 = axes_colorbar(ax[i, 3])
    fig.colorbar(im3, cax=cax3, orientation='vertical')

    i += 1

plt.savefig("rl_sgp_fcsgp_compare_images_1.pdf", format="pdf", bbox_inches="tight", dpi=300)

plt.show()
```
