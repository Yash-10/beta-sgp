```python
rl = pd.read_csv('rl_params_and_metrics.csv')
rl.drop(columns=['Unnamed: 0'], inplace=True)

fcsgp = pd.read_csv('fc_sgp_params_and_metrics.csv')
fcsgp.drop(columns=['Unnamed: 0'], inplace=True)

sgp = pd.read_csv('sgp_params_and_metrics.csv')
sgp.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
success = 0
for i in range(rl.shape[0]):
    if rl.iloc[i]['flux_after'] < rl.iloc[i]['flux_before'] + 0.1 * rl.iloc[i]['flux_before'] and rl.iloc[i]['flux_after'] > rl.iloc[i]['flux_before'] - 0.1 * rl.iloc[i]['flux_before']:
        success += 1
success
```

```python
import matplotlib.pyplot as plt
```

```python
r = rl['flux_after'].to_numpy()
s = sgp['flux_after'].to_numpy()
f = fcsgp['flux_after'].to_numpy()
b = rl['flux_before'].to_numpy()

fig, ax = plt.subplots(1, 1, figsize=(16, 10))
ax.scatter(range(0, len(r)), (b-r)/b, label='RL', facecolors='none', edgecolors='#1C9099', s=70)
ax.scatter(range(0, len(s)), (b-s)/b, label='SGP', c='#636363', marker='x', s=70)
ax.scatter(range(0, len(f)), (b-f)/b, label='FC-SGP', c='#FC9272', marker='+', s=70)
ax.axhline(y=0.0, c='black', linestyle='-', alpha=0.6, label='Perfect flux conservation')
ax.set_xlabel('Image number', fontsize=16)
ax.set_ylabel('Normalized flux residual', fontsize=16)
ax.set_xticks(range(0, 22))
ax.set_ylim([-1.1, +1.1])
ax.legend();
ax.set_title('Flux residual', fontsize=20)

# ax.tick_params(axis='both', which='minor', labelsize=100)

plt.savefig('flux_comparison.png', bbox_inches='tight', dpi=500)
```

```python
import matplotlib.pyplot as plt

r = rl['after_fwhm (pix)'].to_numpy()
s = sgp['after_fwhm (pix)'].to_numpy()
f = fcsgp['after_fwhm (pix)'].to_numpy()
b = rl['before_fwhm (pix)'].to_numpy()

fig, ax = plt.subplots(1, 1, figsize=(16, 10))
ax.scatter(range(0, len(r)), r, label='RL', facecolors='none', edgecolors='#1C9099', s=70)
ax.scatter(range(0, len(s)), s, label='SGP', c='#636363', marker='x',  s=70)
ax.scatter(range(0, len(f)), f, label='FC-SGP', c='#FC9272', marker='+', s=70)
ax.scatter(range(0, len(f)), b, label='Before restoration', c='black', marker='_', s=70)
ax.set_xlabel('Image number', fontsize=16)
ax.set_ylabel('Restored FWHM', fontsize=16)
ax.set_xticks(range(0, 22))
ax.legend();
ax.set_title('FWHM', fontsize=20)
plt.savefig('fwhm_comparison.png', bbox_inches='tight', dpi=500)
```

```python
# jsd divergence plot of radial profile from gaussian fits.
df1 = pd.read_csv('fc_sgp_radprof_params_and_metrics.csv')
df1.drop(columns=["Unnamed: 0"], inplace=True)
df1.columns = ["image", "original_radprof", "orig_fitted_data", "radprof", "fitted_data", "w_before", "w_after", "param_errs"]

df2 = pd.read_csv('sgp_radprof_params_and_metrics.csv')
df2.drop(columns=["Unnamed: 0"], inplace=True)
df2.columns = ["image", "original_radprof", "orig_fitted_data", "radprof", "fitted_data", "w_before", "w_after", "param_errs"]

df3 = pd.read_csv('rl_radprof_params_and_metrics.csv')
df3.drop(columns=["Unnamed: 0"], inplace=True)
df3.columns = ["image", "original_radprof", "orig_fitted_data", "radprof", "fitted_data", "w_before", "w_after", "param_errs"]

# r = rl['after_fwhm (pix)'].iloc[:10].to_numpy()
s = df2['w_after'].to_numpy()
f = df1['w_after'].to_numpy()
r = df3['w_after'].to_numpy()
b = df1['w_before'].to_numpy()

all_max = max(r.max(), s.max(), f.max(), b.max())

fig, ax = plt.subplots(1, 1, figsize=(16, 10))
ax.scatter(range(0, len(r)), r/all_max, label='RL', marker='o', facecolors='none', edgecolors='#1C9099', s=70)
ax.scatter(range(0, len(s)), s/all_max, label='SGP', c='#636363', marker='x', s=70)
ax.scatter(range(0, len(f)), f/all_max, label='FC-SGP', c='#FC9272', marker='+', s=70)
ax.scatter(range(0, len(s)), b/all_max, label='Before restoration', marker='_', c='black', s=70)
ax.set_xlabel('Image number', fontsize=16)
ax.set_ylabel('Normalized 1-Wasserstein distance', fontsize=16)
ax.set_xticks(range(0, 22))
ax.legend();
# ax.set_yscale('log')
ax.set_title('1-Wasserstein distance radial profile', fontsize=20)
plt.savefig('w_comparison.png', bbox_inches='tight', dpi=500)
```
